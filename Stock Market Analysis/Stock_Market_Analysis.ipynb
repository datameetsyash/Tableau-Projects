```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
APPLE= pd.read_csv("SMA_Raw_Data/AAPL.csv", encoding="ISO-8859-1")
FB= pd.read_csv("SMA_Raw_Data/FB.csv", encoding="ISO-8859-1")
GOOGLE= pd.read_csv("SMA_Raw_Data/GOOGL.csv", encoding="ISO-8859-1")
NVIDIA= pd.read_csv("SMA_Raw_Data/NVDA.csv", encoding="ISO-8859-1")
TESLA= pd.read_csv("SMA_Raw_Data/TSLA.csv", encoding="ISO-8859-1")
TWITTER= pd.read_csv("SMA_Raw_Data/TWTR.csv", encoding="ISO-8859-1")
APPLE.sort_index()
FB.sort_index()
GOOGLE.sort_index()
NVIDIA.sort_index()
TESLA.sort_index()
TWITTER.sort_index()
```

```python
dfs= [APPLE,FB,GOOGLE,NVIDIA,TWITTER,TESLA]
```

```python
for i in dfs:
    i["MA50"]= i.Close.rolling(50).mean()
    i["MA200"]= i.Close.rolling(200).mean()
    i["Previous Close"]= i.Close.shift(1)
for i in dfs:
    i["Change In Price"]= i["Close"] - i["Previous Close"]
    i["Percentage Change In Price"]= i["Close"].pct_change()
    i["Previous Volume"]= i["Volume"].shift(1)
for i in dfs:
    i["Change In Volume"]= i["Previous Volume"] - i["Volume"]
    i["Percentage Change In Volume"]= i["Volume"].pct_change()
```
